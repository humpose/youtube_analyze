```python
from bs4 import BeautifulSoup

# 크롤링할 HTML 파일 경로
file_path = "/content/drive/MyDrive/시청 기록.html"

# HTML 파일을 읽어옴
with open(file_path, "r", encoding="utf-8") as file:
    html = file.read()

# HTML을 파싱
soup = BeautifulSoup(html, "html.parser")

# 원하는 요소를 추출하여 출력 또는 다른 처리를 수행
title = soup.title.text
print("페이지 제목:", title)

# 다른 예시: 모든 링크를 추출하여 출력
links = soup.find_all("a")
print("링크 목록:")
for link in links:
    print(link["href"])
```

```python
import pandas as pd
from datetime import datetime
```

```python
# 데이터를 저장할 리스트 초기화
data = []

# 각 행마다 데이터 추출
rows = soup.find_all('div', class_='outer-cell')
for row in rows:
    video_link = row.find('a', href=True)['href']

    channel_links = row.find_all('a', href=True)
    if len(channel_links) > 1:
        channel_link = channel_links[1]['href']
        video_name = row.find('a').text
        channel_name = channel_links[1].text

        br_tags = row.find_all('br')
        if len(br_tags) > 2:
          date_time_text = br_tags[2].next_sibling.strip()



        # 오전/오후 수정
        date_time_text = date_time_text.replace('오전', 'AM')
        date_time_text = date_time_text.replace('오후', 'PM')

        # KST 제거
        date_time_text = date_time_text.replace('KST', '')

        try:
          # 시간 형식으로 변환
          date_time = datetime.strptime(date_time_text, '%Y. %m. %d. %p %I:%M:%S ')
        except ValueError:
          # format에 맞지 않는 경우 행을 건너뜁니다
          continue


        # 데이터를 딕셔너리로 저장
        row_data = {'date_time': date_time,
                    'video_name': video_name,
                    'channel_name': channel_name,
                    'video_link': video_link,
                    'channel_link': channel_link
                    }

        # 데이터를 리스트에 추가
        data.append(row_data)

# 데이터프레임 생성
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd

# 주어진 범위
date_ranges = [
    (pd.to_datetime('2019-01-01').date(), pd.to_datetime('2021-03-02').date()),
    (pd.to_datetime('2021-07-15').date(), pd.to_datetime('2022-08-15').date()),
    (pd.to_datetime('2021-12-30').date(), pd.to_datetime('2022-02-28').date()),
    (pd.to_datetime('2022-06-20').date(), pd.to_datetime('2022-08-31').date()),
    (pd.to_datetime('2022-12-21').date(), pd.to_datetime('2023-02-28').date()),
    (pd.to_datetime('2023-06-20').date(), pd.to_datetime('2023-08-31').date())
]

# 주어진 범위에 해당하는 행 제거
for start_date, end_date in date_ranges:
    df = df[~((df['date_time'].dt.date >= start_date) & (df['date_time'].dt.date <= end_date))]

# 결과 출력
df
```

```python
import re

def convert_duration(duration):
    minutes = re.search(r'(\d+)M', duration)
    seconds = re.search(r'(\d+)S', duration)

    minutes = int(minutes.group(1)) if minutes else 0
    seconds = int(seconds.group(1)) if seconds else 0

    total_seconds = minutes * 60 + seconds
    return total_seconds

sec_df = time_df.copy()

# 'video_time' 열의 값을 초 단위로 변환
sec_df['video_time_seconds'] = sec_df['video_time'].apply(convert_duration)
sec_df
```
