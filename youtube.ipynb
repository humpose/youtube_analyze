```python
import pandas as pd

# pickle 파일 경로 리스트
# 팀원 4명의 유튜브 시청 기록 종합
pickle_files = [
    '/content/drive/MyDrive/Kakao_analysis/youtube_data.pkl',
    '/content/drive/MyDrive/Kakao_analysis/youtube_data2.pkl',
    '/content/drive/MyDrive/Kakao_analysis/youtube_data3.pkl',
    '/content/drive/MyDrive/Kakao_analysis/youtube_data4.pkl'
]

# 빈 DataFrame 생성
combined_df = pd.DataFrame()

# 각 pickle 파일을 순회하며 DataFrame으로 합치기
for file in pickle_files:
    df = pd.read_pickle(file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

combined_df.sort_values('date_time', inplace=True)

# 합쳐진 DataFrame 출력
print(combined_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# pkl 파일 불러오기
#combined_df = pd.read_pickle('/content/drive/MyDrive/Kakao_analysis/youtube_data3.pkl')
combined_df = combined_df.sort_index(ascending=False)

# 날짜 정보와 영상 시청 시간 추출
date_times = combined_df['date_time']
video_times = combined_df['video_time_seconds']

# 초 단위의 시간 차이 계산하여 time_diff 열 추가
time_diff = date_times.diff().dt.total_seconds().abs()

# 비교하여 real_time 값 할당
real_time = pd.Series([min(time_diff[i], video_times[i]) for i in range(len(time_diff))])

# time_diff와 real_time 열 추가
combined_df['time_diff'] = time_diff
combined_df['real_time'] = real_time

# NaN 값이 있는 행 제거
combined_df = combined_df.dropna()

# 날짜 범위 제한
start_date = '2023-03-01'
end_date = '2023-04-26'
combined_df = combined_df[(combined_df['date_time'] >= start_date) & (combined_df['date_time'] <= end_date)]


# 그래프 그리기
plt.plot(combined_df['date_time'], combined_df['real_time'])
plt.xlabel('Date')
plt.ylabel('Real Time (seconds)')
plt.title('Watching YouTube')
plt.xticks(rotation=45)
plt.show()
```
